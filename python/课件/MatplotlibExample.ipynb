## 第一个例子

```python
#绘制折线图
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #正常显示负号
#准备数据
year = [2017, 2018, 2019, 2020]
people = [20, 40, 60, 70]
#生成图表
plt.plot(year, people)
plt.xlabel('年份')
plt.ylabel('人口')
plt.title('人口增长')
#设置纵坐标刻度
plt.yticks([0, 20, 40, 60, 80])
#设置填充选项：参数分别对应横坐标，纵坐标，纵坐标填充起始值，填充颜色
plt.fill_between(year, people, 20, color = 'green')
#显示图表
plt.show()
```

# 综合实例

## 离散型数据可视化

通过数据可视化能够清晰地表达离散型变量的分布特征，例如统计某游戏用户的性别比例、各年龄段消费者的消费能力差异、某产品在不同地区销售情况等。对于这样的离散型变量统计描述，可以使用饼图、条形图等对其进行展示。
前文已经介绍了饼图、条形图等图形的绘制方法，但每种图形似乎只能反映单个离散变量的统计结果，如果想传递多个离散变量的信息可借助堆叠图、分块图实现。堆叠图利用横坐标代表一个维度的离散变量，用堆叠起来的“块”代表另一个维度的离散变量，其最大优点是可以方便比较累积和，其缺点是不易区分“块”之间的差异。分块图是将堆叠的“块”水平排开，这样方便对比另一个维度的数据。
下面以胡润财富榜中亿元人民币资产高净值家庭数排名前五的城市数据为例绘制分块图，具体代码如下：

```python
import  matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
#准备数据
#城市名称
cities = ['北京','上海','香港','深圳','广州']
#2020年各城市亿元人民币资产家庭数
counts2020 = [19300,16200,12500,5590,4350]
#2021年各城市亿元人民币资产家庭数
counts2021 = [19700,16800,11700,5900,4500]
#创建画布和坐标系
plt.figure(figsize=(8,4), dpi=80)
ax = plt.subplot(111)
#设置属性
ax.set_title('近两年5个城市亿元人民币资产家庭数比较')
ax.set_ylabel('亿元人民币资产家庭数')
# 绘制分块图
barwidth = 0.4
x = np.arange(len(cities))
ax.bar(x-barwidth/2 ,counts2020,label='2020',width=barwidth)
ax.bar(x+barwidth/2 ,counts2021,label='2021',width=barwidth)
ax.set_xticks(np.arange(len(cities)))
ax.set_xticklabels(cities)
ax.legend()
# 显示图像
plt.show()
```

## 数值型数据可视化

数据分析与处理中的大量数据都是数值型变量，对数据进行探索和分析时，可视化是最佳手段。利用matplotlib绘制线性图、直方图、箱线图等能够直观地观察数值型数据的分布，发现数据的特征。例如不同区域房价变化特点、视频号每日流量趋势、客户消费数据变化情况等。对于连续的数值型数据的统计描述，可以使用线性图、直方图、箱线图、密度图等对其进行展示，下面以某地区近年月平均气温变化趋势为例绘制线性图。
某地区近年月平均气温数据存储在磁盘文件temp.csv中，数据格式如下表所示：

<img src= './Table4.2.png' width = "50%">

```python
import  matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
#准备数据，数据存储在temp.csv文件中
temp = pd.read_csv(r'temp.csv')
#创建画布和坐标系
plt.figure(figsize=(8,4), dpi=80)
ax = plt.subplot(111)
#设置属性
ax.grid(visible=True,linestyle = "--",color = "gray", linewidth = "0.5")
ax.set_title('某地区近年月平均气温波动趋势')
ax.set_xlabel('月份')
ax.set_xlim(0,13)
ax.set_xticks(range(1,13))
ax.set_ylabel('气温（℃）')
# 绘制散点图
ax.plot(temp.month,temp.y2021,linestyle='-',marker='o',label='2021年')
ax.plot(temp.month,temp.y2020,linestyle='--',marker='.',label='2020年')
ax.plot(temp.month,temp.y2019,linestyle='-.',marker='v',label='2019年')
ax.legend()
# 显示图像
plt.show()
```

## 关系型数据可视化

在众多的可视化图形中，有一类图形专门用于探究两个或两个以上变量之间的关系。例如探索商品类别和客户年龄之间的关系、商品销售额与销售时间的关系等。如果需要研究两个数值型变量之间是否存在某种线性关系，或趋势性的非线性关系，或聚类关系等，散点图是最佳选择。
下面以鸢尾花数据集为例，探究如何应用scatter函数绘制鸢尾花花瓣长度与宽度的散点图，以找到鸢尾花花瓣长度与宽度之间的关系。

<img src= './iris.png' width = "60%">

```python
import  matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
#准备数据，数据存储在d:\python\data\iris.csv文件中
iris = pd.read_csv(r'iris.csv')
df=pd.DataFrame(iris)                     #转化为dataframe数据类型
display(df)
df.columns = ['SepalLen','SepalWidth','PetalLen','PetalWidth','Class']           #花瓣长度，花萼长度两个特征
display(df)
#创建画布和坐标系
plt.figure(figsize=(8,4), dpi=80)
ax = plt.subplot(111)
#设置属性
ax.grid(visible=True,linestyle = "--",color = "gray", linewidth = "0.5")
ax.set_title('鸢尾花花瓣长度与宽度关系')
ax.set_xlabel('花瓣宽度（cm）')
ax.set_ylabel('花瓣长度（cm）')
# 绘制散点图
ax.scatter(x = df['PetalWidth'], #指定散点图的x轴数据
          y = df['PetalLen']  #指定散点图的y轴数据
          )
# 显示图像
plt.show()
```

# 绘制3D图形

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
import matplotlib as mpl
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits.mplot3d.art3d import Poly3DCollection,Line3DCollection
plt.ion()
fig = plt.figure()

for i in range(10000):

    x = np.linspace(1,20,9)
    y = np.arange(10,19,1)
    z = np.random.randint(20,50,9)  # numpy分别生成三个维度数据

    ax = fig.add_subplot(111, projection="3d") #通过在add_subplot(111,projection='3d'）将子图坐标修改成三维
    ax.set_xlim3d(0, 20)  # 指定x轴坐标值范围
    ax.set_ylim3d(0, 20)  # 指定y轴坐标值范围
    ax.set_zlim3d(0, 50)  # 指定z轴坐标值范围
    ax.plot(x,y,z,'bo--')  # 参数与二维折现图不同的在于多了一个Z轴的数据
    plt.show()
    plt.pause(0.3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def randrange(n, randFloor, randCeil):
    rnd = np.random.rand(n)                   #生成n个随机数（值0~1）
    return (randCeil-randFloor) * rnd + randFloor  #生成n个vmin~vmax的随机数

plt.rcParams['font.sans-serif']=['SimHei']    #显示中文 
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")  #添加子3D坐标轴,
n = 100
for zmin, zmax, c, m, l in [(4, 15, 'r', 'o', '低值'),(13, 40, 'g', '*', '高值')]:  #形状和颜色
    x = randrange(n, 0, 20)
    y = randrange(n, 0, 20)
    z = randrange(n, zmin, zmax)
    ax.scatter(x, y, z, c=c, marker=m, label=l, s=z * 6)

ax.set_xlabel("X-value")
ax.set_ylabel("Y-value")
ax.set_zlabel("Z-value")
ax.set_title("高/低值3D散点图", alpha=0.6, size=15, weight='bold') 
ax.legend(loc="upper left")              #图例位于左上角

plt.show()
```
