具体如图：

SVD 奇异值分解是在机器学习领域广泛应用的算法，它不光可以用于降维算法中的特征分解，还可以用于推荐系统，以及自然语言处理等领域。是很多机器学习算法的基石。下面将讨论 SVD 在主
成分分析 PCA(principle component analysis)方法中是如何运用的。
**PCA 算法简介**：PCA 在机器学习中是一种常用的无监督学习算法，通常用于高维数据集的探索与可视化，还可以用作数据压缩和预处理等。它通过构建一种被称为主成分的变量，把具有相关性的高维变量合成为线性无关的低维变量，从而能够减少数据的维度。主成分能够尽可能保留原始数据的信息。  
PCA 的算法过程如下：

可以看出，PCA 算法的核心步骤是第 3 步：协方差矩阵$X^TX$的特 征 值 分 解 。 为 了 说 明二者的关系,我们对$A_{m \times n}=U_{m \times m}\sum_{m \times m}V_{n \times n}^T $做如下的变换：  
$$A^TA=(U_{m \times m}\sum_{m\times m}V_{n\times n}^T)^T(U_{m \times m}\sum_{m\times m}V_{n\times n}^T)=V_{n\times n}\sum_{m\times m}^2V_{n\times n}^T$$

所以 V 是 $A^TA$ 特征值分解的特征向量按照列组成的正交矩阵，$\sum^2$是$A^TA$特征值组成的对角矩阵, 也可以看出$A_{m \times n}$的奇异值$\sigma_i$
是 $A^TA$ 特征值 $\lambda_i$ 的平方根。即：
$$\sigma_i=\sqrt{\lambda_i}$$
假设 $A^TA$ 的特征向量为$V_i$，则矩阵$U_{m\times m}$中的元素$U_i=\frac{Av_i}{\sigma_i}$。  
可见，SVD 的关键也是在于对矩阵 $A^TA$ 的特征值分解。很明显二者所解决的问题非常相似，都是对一个实对称矩阵进行特征值分解。如果令 $X=A^T$，则$XX^T = A^TA$，因此 PCA 算法中协方差矩阵 $X^XT$ 的特征值分解可以转换为矩阵 A 的 SVD 问题求解。
